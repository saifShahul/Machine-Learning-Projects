```python
column=[i for i in range(0,66)]
train_dataset=pd.read_csv("/kaggle/input/handwritten-digit-identification/train.csv",names=column)
train_dataset
test_dataset=pd.read_csv("/kaggle/input/handwritten-digit-identification/test.csv",names=column[:65],index_col=0)
test_dataset
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X=train_dataset.iloc[:,1:65]
Y=train_dataset.iloc[:,65]
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y , random_state=42)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
mlp_mod = MLPClassifier(hidden_layer_sizes=(257,269), max_iter = 300 ,  random_state=1)
mlp_mod.fit(X_train, Y_train)

y_pred = mlp_mod.predict(X_test)
print(mlp_mod.score(X_train,Y_train))
print(mlp_mod.score(X_test,Y_test))
```

```python
mlp_mod.fit(X,Y)
my_predict = mlp_mod.predict(test_dataset)
```

```python
output= pd.DataFrame({'ID': test_dataset.index, 'Category': my_predict})
output.to_csv('sample_submission.csv',index=False)
output
```
