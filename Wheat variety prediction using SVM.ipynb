```python
import matplotlib.pyplot as plt
test_data=pd.read_csv("/kaggle/input/wheet-variety-prediction-svm-b2/wheet_test.csv")
test_data
```

```python
train_data=pd.read_csv("/kaggle/input/wheet-variety-prediction-svm-b2/wheet_train.csv")
train_data
```

```python
X = train_data.iloc[:,0:8].values
y = train_data.iloc[:,8].values
X
y
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=0)
X_train
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear',gamma='auto', C=2)
classifier.fit(X_train, y_train)
```

```python

y_pred = classifier.predict(X_test)
y_pred
newid=X_test[:,0]
```

```python
output=pd.DataFrame({'ID':newid,'Type':y_pred})
output.to_csv('submission.csv',index=False)
output
```
